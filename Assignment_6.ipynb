# b. Deduce the percentage of developers who know python in each country.

```python
schema['Country'].nunique()
```

```python
language = schema[schema['LanguageWorkedWith'].notnull()]
```

```python
language.head()
```

```python
countries = language['Country'].unique()
```

```python
country_python = {}
```

```python
language.groupby(['Country'])['LanguageWorkedWith'].value_counts()
```

```python
for country in countries:
    for i in language.groupby(['Country'])['LanguageWorkedWith'].value_counts()[country].index:
        l = []
        l.append(i.split(';'))
        # print(l)
        for j in l:
            if 'Python' in j:
                if country not in country_python.keys():
                    country_python[country] = 1
                else:
                    country_python[country] += 1
                       
```

```python
country_python
```

```python
def plot_dimension_count(unique_dim_dict, plot_title):
    dim_count = pd.DataFrame.from_dict(unique_dim_dict, orient='index', dtype=None)
    dim_count.columns = ['Count']
    dim_count.sort_values('Count', ascending=False, inplace=True)
    
    dim_count[:10].plot(kind = 'pie', figsize=(12,12), fontsize = 10, title = plot_title, subplots = True, autopct='%1.1f%%');
    print(dim_count[:10])
```

```python
plot_dimension_count(country_python,'The Most Popular Language')
```

# c.  Generate a report for the average salary of developer based on continent.

```python
import pycountry_convert as pc
```

```python
def country_to_continent (country_name): 
    if country_name != 'Other Country (Not Listed Above)':
 #           print(country_name)
        try:
            country_alpha2 = pc.country_name_to_country_alpha2 (country_name) 
 #                print(country_alpha2)

            country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
 #                    print(country_continent_code) 
            country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code) 
            return country_continent_name
        except:
            return "unknown"
    else:
        return "unknown"
```

```python
schema['Country'].replace({'Hong Kong (5.A.R.)': 'Hong Kong'}, inplace =True)
schema['Country'].replace({ 'Libyan Arab Jamahiriya': 'Libya'},inplace =True)
schema['Country'].replace({'Venezuela, Bolivarian Republic of...': 'Venezuela'}, inplace =True) 
schema['Country'].replace({'The former Yugoslav Republic of Macedonia': 'North Macedonia'}, inplace =True)
schema['Country'].replace({'Republic of Korea': 'South Korea'}, inplace =True)
schema['Country'].replace({ 'Congo, Republic of the...': 'Republic of the Congo'}, inplace =True)
schema['Country'].replace({'Timor-Leste': 'East Timor'}, inplace =True)
```

```python
names = [country_to_continent(i) for i in schema['Country']]
schema['Continent'] = names
```

```python
salary_notnull = schema[schema['ConvertedComp'].notnull()]
```

```python
a = salary_notnull.groupby(['Continent'])['ConvertedComp'].mean().sort_values(ascending = False)
```

```python
plt.figure(figsize=(16,9))
plt.title('Salary countplot based on Continent', fontsize=20, fontweight='bold', y=1.05,)
plt.ylabel('Salary', fontsize=15) 
plt.xlabel('Continent', fontsize=15)
a.plot(kind='bar')
```
